```python
import random

class Account:
    # Account 클래스로부터 생성된 계좌 객체의 개수를 저장
    account_count = 0

    def __init__(self,name,balance):
        # 은행이름, 예금주, 계좌번호, 잔액이 설정
        self.bank = "SC은행" # 은행이름
        self.name = name     # 이름
        self.balance = balance #  잔액

        # num1 = random.randint(0, 999) # 3자리 랜덤
        # num2 = random.randint(0, 99)  # 2자리 랜덤
        # num3 = random.randint(0, 999999)  # 6자리 랜덤
        # TypeError: unsupported operand type(s) for +: 'int' and 'str'
        # num1 = str(num1).zfill(3)      # 1 -> '1' -> '001'
        # num2 = str(num2).zfill(2)      # 1 -> '1' -> '01'
        # num3 = str(num3).zfill(6)      # 1 -> '1' -> '0000001'
        # self.account_number = num1 + '-' + num2 + '-' + num3  # 001-01-000001

        num1 = str(random.randint(0, 999)).zfill(3)  # 3자리 수
        num2 = str(random.randint(0, 99)).zfill(2)    # 2자리 수
        num3 = str(random.randint(0, 999999)).zfill(6) # 6자리 수
        self.account_number = num1 + '-' + num2 + '-' + num3 # 계좌번호



        # account_count += 1 # 계좌 객체의 개수 증가
        # nboundLocalError: local variable 'account_count' referenced before assignment
        Account.account_count += 1 # 계좌 객체의 개수 증가

        self.deposit_count  = 0 # 입금 횟수 초기화
        self.deposit_details_list = []  # 입금 내역 초기화
        self.withdraw_details_list = [] # 출금 내역 초기화

        print(f"__init__ 은행이름 :{self.bank}, 예금주 : {self.name}, 계좌번호 : {self.account_number}, 잔고 : {self.balance:,} 원 ")


    # 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가
    @classmethod
    def get_account_num(cls):
       print(f"계좌의 개수 :{cls.account_count}")


    # 입금 메서드
    def deposit(self , amount):

      # 입금은 최소 1원 이상만 가능
      if amount >= 1 :

        self.balance += amount # 입금액
        self.deposit_details_list.append(amount) # 입금 내역에 추가
        self.deposit_count += 1
        print(f"self.deposit_count: {self.deposit_count}")
        print(f"self.deposit_details_list: {self.deposit_details_list}")
        print(f"self.balance: {self.balance}")

        # 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경
        if self.deposit_count % 5 == 0 :
          print(f"이자지급액 전 self.balance: {self.balance}")
          print(f"이자지급액 : {self.balance * 0.01} ")
          self.balance += (self.balance * 0.01)
          print(f"이자지급 후 잔고 : {self.balance} ")
      else  :

        print("입금은 최소 1원 이상만 가능합니다")


    # 출금 메서드
    def withdraw(self, amount):
      print(f"withdraw- self.balance: {self.balance}")
      print(f"withdraw- amount: {amount}")
      # 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.
      if self.balance >= amount :

        self.balance -= amount # 출금액 차감
        self.withdraw_details_list.append(amount) # 출금 내역에 추가
        print(f"self.withdraw_details_list: {self.withdraw_details_list}")

      else :
        print("출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.")

    # 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요. (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
    def display_info(self):

      # (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
      # 잔고는 세자리마다 쉼표를 출력 ★
      print(f"은행이름 :{self.bank}, 예금주 : {self.name}, 계좌번호 : {self.account_number}, 잔고 : {self.balance:,} 원 ")
      #2. f-string 사용
      #Python 3.6 이상에서는 f-string 내에서 {}를 사용해 직접 포맷을 지정할 수 있습니다. 숫자 포매팅을 위해 : 이후에 쉼표(,)를 추가함으로써 천 단위 구분자를 적용할 수 있습니다.

    # 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요. (입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)
    # 입금 내역 출력 메서드
    def deposit_history(self):

      print("입금내역")
      for amount in self.deposit_details_list:
        print(f"{amount} 원")

    # 출금 내역 출력 메서드
    def withdraw_history(self):

      print("출금내역")
      for amount in self.withdraw_details_list :
        print(f"{amount} 원")

# 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장
accounts = [
  Account("이한나", 50000000),
  Account("변우석", 350000),
  Account("차은우", 10000000),
  Account("아이펠", 220000)
]

# 첫 번째 계좌에 대한 입금 5번과 출금 2회
account = accounts[0]
account.deposit(500000)
account.withdraw(150000)
account.deposit_history()
account.withdraw_history()
account.deposit(30000)
account.deposit_history()
account.deposit(20000)
account.deposit_history()
account.deposit(10000)
account.deposit_history()
account.deposit(60000)
account.deposit_history()
account.withdraw(60000000)
account.withdraw_history()

# 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.
for account in accounts:

  if account.balance >= 1000000 :

    # Account.display_info()
    # TypeError: Account.display_info() missing 1 required positional argument: 'self'
    # Account.display_info() 메서드를 호출하려고 하지만, 여기서 올바르게 호출하지 않고 있습니다.
    # 이 메서드는 인스턴스 메서드이기 때문에 특정 Account 인스턴스를 통해 호출되어야 합니다.
    # Account.display_info() 형식은 클래스 메서드나 정적 메서드를 호출할 때 사용되는 형식입니다.
    # 인스턴스 메서드는 개별 객체의 데이터를 사용하므로, 반드시 객체를 통해 호출되어야 합니다.
    # 예제에서는 account 객체가 각각의 계좌 인스턴스를 나타내므로,
    # 각 계좌의 정보를 출력하려면 해당 객체의 display_info() 메서드를 직접 호출해야 합니다.
    print("잔고가 100만원 이상인 고객의 정보")
    account.display_info()




```
