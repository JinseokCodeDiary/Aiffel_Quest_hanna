아이펠 캠퍼스에서는 그루들에게 학번을 부여하고 퀘스트 점수를 관리하고 있습니다. 이제부터 그루들의 정보와 학번과 점수를 관리할 수 있는 Aiffel 클래스를 만들어보도록 하겠습니다.

Q1. Aiffel 클래스 만들기
생성자에는 이름, 입학연도, 기수, 과정명의 인자를 받습니다. 이를 통해 학번을 생성하고 학번 정보로 8기 코어 그루 명단을 관리해봅시다.

학번은 7자리로 이루어져 있습니다. : YYNNCXX  
● 학번 생성 규칙:  
○ YY: 연도 마지막 두자리 (2024 -> 24)  
○ NN: 기수 (온라인 8기는 31기)  
○ C: 과정 종류 (리서치 = 1, 코어 = 2)  
○ XX: 가나다순 번호  

Q2. 점수 관리 메서드
다음과 같은 기능이 추가되어야 합니다.
● 퀘스트 점수 부여
● 퀘스트 점수 차감 (페널티 부여)
● 현재 총점 출력

Q3. 그루 명단 관리하기  
다음과 같은 기능을 가진 프로그램을 만들어 봅시다.  
● 전체 명단과 정보를 보는 기능  
● 퀘스트 점수를 기준으로 내림차순으로 정렬해서 보는 기능  
● 특정 인물 정보를 삭제하는 기능  


Q4. 그룹핑 기능 추가하기
그루 명단을 기준으로 2인, 3인, 4인 조를 랜덤하게 구성하는 기능을 추가해봅시다.




```python
import random

class Aiffel:

  all_students = []  # 모든 학생 객체를 저장할 클래스 변수
  registr_counter = {}  # 기수와 과정별 학생 수를 카운팅하는 딕셔너리

  def __init__(self, name, adm_yr, crs_num, crs_name):

    #  이름, 입학연도, 기수, 과정명
    self.name = name          # 이름
    self.adm_yr = adm_yr      # 입학연도
    self.crs_num = crs_num    # 기수
    self.crs_name = crs_name  # 과정명
    self.scores   =  0
    self.penalties   =  0
    self.std_id   =  self.make_std_id()
    Aiffel.all_students.append(self)  # 객체 생성 시 목록에 추가

  def make_std_id(self) :

    # 학번은 7자리로 이루어져 있습니다. : YYNNCXX
    # ● 학번 생성 규칙:
    # ○ YY: 연도 마지막 두자리 (2024 -> 24)
    # ○ NN: 기수 (온라인 8기는 31기)
    # ○ C: 과정 종류 (리서치 = 1, 코어 = 2)
    # ○ XX: 가나다순 번호

    #과정 종류 셋팅
    crs_type = {'리서치': 1 , '코어': 2 }
    crs_num_mapping = {8:31} # 온라인 8기는 31기

    ##YYNNCXX
    yy = str(self.adm_yr)[-2:] # 연도 마지막 두자리 (2024 -> 24)
    # print(f"map_num_crs_num: {crs_num_mapping.get(self.crs_num)}" )
    nn = str(crs_num_mapping.get(self.crs_num))
    c = str(crs_type.get(self.crs_name, '0'))
    # 기수와 과정 조합 키 생성
    key = yy + nn + c
    print(key)
    # 해당 키에 대한 가나다순 번호(일련번호)를 증가..  ★
    Aiffel.registr_counter[key] = Aiffel.registr_counter.get(key, 0) + 1
    xx = str(Aiffel.registr_counter[key]).zfill(2)

    print(self.name)
    print(xx)
    return yy + nn + c + xx

  # 점수 계산
  def add_score(self, score) :

      self.scores += score

  # 페널티 계산
  def add_penalty(self, penalty) :

      self.penalties += penalty

  # 점수 계산
  def display_score(self) :

      total_score = self.scores - self.penalties
      print(f"학번 {self.std_id} {self.name}'s 총점: {total_score} (점수: {self.scores}, 페널티: {self.penalties})")

  @classmethod
  def display_students(cls, students=None):
    if students is None:
        students = cls.all_students
    for student in students:
        student.display_score()


  # 점수 별 내림차순
  @classmethod
  def sort_students(cls):

      sorted_students = sorted(cls.all_students, key=lambda x: x.scores - x.penalties, reverse=True)
      cls.display_students(sorted_students)
      print("sort_students")

  # 학생 삭제
  @classmethod
  def remove_student(cls, name):
      cls.all_students = [student for student in cls.all_students if student.name != name]


# 임시 데이터 예시
students = [
    Aiffel("강대식", 2024, 8, "코어"),
    Aiffel("김기홍", 2024, 8, "코어"),
    Aiffel("박진석", 2024, 8, "코어"),
    Aiffel("신승우", 2024, 8, "코어"),
    Aiffel("이한나", 2024, 8, "코어")
]


# 점수 및 페널티 부여
students[0].add_score(44)
students[0].add_penalty(22)
students[1].add_score(37)
students[1].add_penalty(9)
students[2].add_score(27)
students[2].add_penalty(22)
students[3].add_score(48)
students[3].add_penalty(24)
students[4].add_score(46)
students[4].add_penalty(20)

# 점수 출력
Aiffel.display_students()

# 특정 학생 삭제
Aiffel.remove_student("이한나")

# 이한나 학생 퇴학 후 학생 목록 출력
print("이한나 학생 퇴학 후 현재 학생 목록:")
Aiffel.display_students()

# 학생 점수 정렬 및 출력
print("점수 순으로 학생 목록 정렬 및 출력:")
Aiffel.sort_students()
Aiffel.display_students()

```
